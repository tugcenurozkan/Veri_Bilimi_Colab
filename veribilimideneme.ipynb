**KLASİK KÜME İŞLEMLERİ**

> difference() ->> ile iki kümenin farkını ya da "-" ifadesi

> intersection() ->> iki küme kesişimi ya da "&" ifadesi

> union() ->> iki kümenin birleşimi

> symmetric_difference() ->> ikisinde de olmayanları




```
set1=set([1,3,5])
set2=set([1,2,3])
```

```
set1.difference(set2)
#set1 de olup set2 de olmayan
```

```
set1-set2
#bu işlem yukarda belirtildiği gibi "-" ifadesiyle de sağlanmış olur
#tabiki "-" işlemi yapmak toplama çarpma gibi işlemlerin yapıldığı anlamına gelmez.
```

```
set1.symmetric_difference(set2)
#hem set1 de hem set2 de aynı anda bulunmayanları temsil eder.
```

```
set1.intersection(set2)
#kesişim işlemi hem böyle hem de aşağıdaki "&" ifadesi ile sağlanmaktadır.
```

```
set1 & set2
```

```
set1.union(set2)
#unique olarak birleşmiş oldu.
```

```
set1.intersection_update(set2)
#ile set1 ve set2 nin kesişimini set1 in kalıcı elemanları olmasını sağlamak için kullanılır.
```

```
set1
```

**İNTERNETTEN BULUNAN ARA ÖRNEKLER**

```
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def introduceyourself(self):
        print("My name is " + self.name)
        print("My age is " + str(self.age))

        
class Teacher(Person): #this class inherits the class above!
    def stateprofession(self):
        print("I am a teacher!")
```

```
author = Teacher("Maarten",30)
author.introduceyourself()
author.stateprofession()
```

```
class Teacher(Person): #this class inherits the class above!
    def __init__(self, name, age):
        self.courses = [] #initialise a new variable
        super().__init__(name,age) #call the init of Person
        
    def stateprofession(self):
        print("I am a teacher!")        
    
    def introduceyourself(self):
        super().introduceyourself() #call the introduceyourself() of the Person
        self.stateprofession()
        print("I teach " + str(self.nrofcourses()) + " course(s)")
        for course in self.courses:
            print("I teach " + course)
      
        
    
    def addcourse(self, course):
        self.courses.append(course)
        
    def nrofcourses(self):
        return len(self.courses)
    
    
author = Teacher("Maarten",30)
author.addcourse("Python")
author.introduceyourself()
```

```
class LineCounter:
  def __init__(self,filename):
    self.file=open(filename,'r')
    self.lines=[]
  def read(self):
    self.lines=[line for line in self.file]
  def count(self):
    return len(self.lines)
```

**İleri Aggregate İşlemleri(Aggregate, Filter,Transform,Apply)**

```
import numpy as np
import pandas as pd
df=pd.DataFrame({"gruplar":["A","B","C","A","B","C"],
                 "degisken1":[12,23,63,45,56,21],
                 "degisken2":[100,236,245,654,169,785]},
                columns=["gruplar","degisken1","degisken2"])
df
```

**Aggregate**

```
df.groupby("gruplar")
```

```
?df.aggregate()
```

```
df.groupby("gruplar").aggregate(["min",np.median,max,np.mean]) #median burda np özeliiği olduğu için np.median olarak yazdık. Mesela  burda np.median veya np.mean yazmayınca tanımlı değil diyerek hata veriyor.
#burda birden fazla aggregate fonksiyonu kullanılmak istenirse diye bu şekilde yazmak mümkün 
```

**NOT: Pandas da fonksiyonları "" içine veya tırnak işareti olmadan yazasan da algılıyor ancak np.median için aynı şey geçerli değil "" içine alınca hata verir**

```
df.groupby("gruplar").agg({"degisken1":[min,np.median],"degisken2":[max,np.mean]}) 
#değişkenler için farklı hesaplamalar yapmak istersek dictionary yapısı ile ayrı ayrı belirtmek mümkün 
#agg ->>aggregate ile aynı anlamda sadece kısaltması 
```

```
df.groupby("gruplar").agg({"degisken1":min,"degisken2":"max"}) 
#min mmax ı [] içine yazmayınca isimleri gözükmüyor farklı olarak 
```

**Filter**

```
def filter_function(x):                         #pandas içerisinde direk olarak dataframe içindeki bir eleman üzerinde işlem yapma fonksiyonu olmasığı için kendimiz yazarız
  return x["degisken1"].std()>24
```

```
df.groupby("gruplar").std()
```

**Confidence Interval(Güven Aralığı)**
> Anakütle parametresinin tahmini değerini kapsayabilecek iki sayıdan oluşan bir aralık bulunmasıdır.

**Güven Aralığını kendimiz hesaplamak için**
1. n, ortalama, standart sapma yı bul
> n=100, ortalama=180, standart sapma=40
2. Güven aralığına karar ver: 95 mi 99 mu? : Default olarak %95 kabul edilir.
> Z tablo değerini hesapla(1,96-2,57) :Klasik bilinen Z tablosundan 95 e karşılık değer değeri buluyoruz o da 1,96 dır.
3. Yukardaki değerleri kullanarak değer aralığını hesapla:
> ortalama-+z(standart sapma/√n sayısı) => 180-+1,96(40/√100)

Sonuç: 180-+7,84 yani 172 ile 188 aralığındadır diyebiliriz.

**Büyük Sayılar Yasası**
> Bir rassal değişkenin uzun vadeli kararlılığını tanımlayan olasılık teoremidir.

Yukardaki tanımı açıklayacak olursak eğer : Normalde yazı tura atıldığında yazı veya tura gelme olasılığına %50 deriz. Bir deneyle devam edelim diyelim ki 10 kere para attın ve 3 kere yazı geldi sadece ve yazı gelme oranı %30 oldu. Bu değer bizim ön gördüğümüz sayı(%50) değil. **Büyük Sayılar Yasası** ile neden %50 dendiğini daha iyi anlıyoruz çünkü deney yapma sayısı arttıkça yazı ve tura gelme olasılıklarının birbirine yaklaştığı görülmekte.

> Aşağıdaki kod satırı ile bunu kanıtlamak mümkün

```
  import numpy as np
  rand=np.random.RandomState(123) #RandomState ->> seed in yaptığını yapıyor aslında ve random atanan değerlerin değişmemesini sağlıyor
  for i in np.arange(1,21):  #1 den 21 e kadar gezinsin
    deney_sayisi=2**i
    yazi_turalar=rand.randint(0,2,size=deney_sayisi)  #rand ile random atamalrın değişmiyceği bilgisi sabit kalsın isteniyor.
    yazi_olasiliklari=np.mean(yazi_turalar)
    print("Atış sayısı: ",deney_sayisi,"----","Yazi olasiligi: %.2f" %(yazi_olasiliklari*100)) #%.2f ise . dan sonra 2 absamak ondalık olsun der
#görüldüğü üzere satış sayısı artmasıyla değerler birbirine yaklaşıyor.
```

**Binom Dağılımı**

> Bağımsız n deneme sonucu k başarılı olma olasılığı ile ilgilenildiğinde kullanılan dağılımdır. Bernoulli den farkı n defa tekrar edilmesi.

![My image](https://github.com/tugcenurozkan/PhotosForColab/blob/main/binomformul.png?raw=true)

```
from scipy.stats import binom
```

```
#bir reklemın 100 kişiden kaç ı tıklar hesabı yapılıyor. 100 kişiden 1 kişi, 5kişi, 10 kişi tıklama olasılıkları
p=0.01  #0,01 başarma oranı
n=100   #kaç kere denendiği
rv=binom(n,p)
print(rv.pmf(1)) #1 kişi tıklarsa
print(rv.pmf(5)) #5 kişi tıklarsa
print(rv.pmf(10)) #10 kişi tıklarsa
```

**Poisson Dağılımı**
> Belirli bir zaman aralığında belirli bir alanda nadiren rastlanan olayların olasılıklarını hesaplamak için kullanılır.

> Ortalama ve varyans aynıdır. Lamdaya eşittiir ve lambda beklene değerlerin ortalamasını verir.

>n>50, n*p<5 şartını sağlaması nadir olaydır. 

>Rassal denemeler iki sonuçlu olmalı ve aynı koşullar altında gerçekleşmelidir.Aynı zmanada rassal değişkenler birbirinden bağımsız olmalıdır.

![My image](https://github.com/tugcenurozkan/PhotosForColab/blob/main/poissonformul.jpeg?raw=true)

```
from scipy.stats import poisson
#bir ilan giriş hata olasılıkları
lambda_=0.1
rv=poisson(mu=lambda_)
print(rv.pmf(k=0))  #hiç hata olmaması 
print(rv.pmf(k=3))   #3 tane hata olma olasılığı
print(rv.pmf(k=5))
```

**Normal Dağılım**(Sürekli Olasılık Dağılımı)
> Normal dağıldığı bilinen sürekli rassal değişkenler için olasılık hesabı

> μ bize ortamalayı verir.(Beklenen değer)-----
σ standart sapmayı -----
σ^2 ise varyansı

![My image](https://github.com/tugcenurozkan/PhotosForColab/blob/main/normalda%C4%9F%C4%B1l%C4%B1mformul.jpeg?raw=true)

```
#satış olasılıklarının hesaplanması
#ortalama 80 verilmiş standart sapma ise 5 verilmiş
from scipy.stats import norm
#cdf=cumulative density function 
1-norm.cdf(90,80,5) #90 dan fazla olma olasılığı
```

```
1-norm.cdf(70,80,5) #70 den fazla olma olasılığı
```

```
norm.cdf(73,80,5) #73 den az olma olasılığı *burda 1 den çıkarma işlemi yok artık normal dağılım grafiğini göz önüne getir.
```

```
norm.cdf(90,80,5)-norm.cdf(85,80,5) #85 ile 90 arasında olması *integral işlemi yapıyorsun burda aslında
```
